```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
x = []
y = []

with open('log_4_withtest_512.txt','r') as f:
    for i in f.readlines():
        i = i.rstrip()
        x.append(int(i.split(' ')[0].split(':')[1]))
        y.append(float(i.split(' ')[1].split(':')[1]))
```

```python

plt.figure()
plt.plot(x[2::], y[2::], linewidth=1,linestyle="solid" ,label='tra')#'bo-'表示蓝色实线，数据点实心原点标注
## plot中参数的含义分别是横轴值，纵轴值，线的形状（'s'方块,'o'实心圆点，'*'五角星   ...，颜色，透明度,线的宽度和标签 ，
# plt.xlim(0, 300)  # x轴坐标范围
# 
plt.xlabel('epoch') #x_label
plt.ylabel('loss')#y_label
plt.legend()  #显示上面的label 
plt.ylim(0,1)#仅设置y轴坐标范围
plt.show()
```
